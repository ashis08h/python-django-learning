```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('data_folder/weather_pivot.csv')
```

```python
df
```

```python
# code to show dataframe having day as index and city as column
# below code will display temprature data as well as humidity also.
df.pivot(index='day', columns='city')
```

```python
# code to pass value also with pivot table
df.pivot(index='day', columns='city', values='humidity')
```

```python
# code to do pivot having index as humidity and column as city
df.pivot(index='humidity', columns='city')
```

```python
df1 = pd.read_csv('data_folder/weather_pivot2.csv')
```

```python
df1
```

```python
# code to get aggregate concept such as avg based on city since we have multiple value for same city and same date.
df1.pivot_table(index='city', columns='day')
```

```python
df1
```

```python
gf = df1.groupby(['city', 'day'])
```

```python
for item, value in gf:
    print(item)
    print(value)
```

```python
gf.describe()
```

```python
# code to assign custom aggfunc to pivot table
df1.pivot_table(index='city', columns='day', aggfunc='sum')
```

```python
# code to show count with pivot table
df1.pivot_table(index='city', columns='day', aggfunc='count')
```

```python
# code to apply margins with pivot table
df1.pivot_table(index='city', columns='day', margins=True)
```

```python
df3 = pd.read_csv('data_folder/weather_pivot3.csv')
```

```python
df3
```

```python
df3['date'] = pd.to_datetime(df3['date'])
```

```python
df3
```

```python
# code to get avergage value monthwise 
df3.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city')
```

```python
# samething
df3.pivot_table(index=pd.Grouper(freq='ME', key='date'), columns='city')
```
