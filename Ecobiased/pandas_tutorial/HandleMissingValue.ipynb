```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_folder/new_weather.csv')
```

```python
df
```

```python
# code to replace -9999 with NaN.
new_df = df.replace(-99999, np.NaN)
```

```python
new_df
```

```python
# code to replace multiple special value with NaN.
new_df = df.replace([-99999, -88888], np.NaN)
```

```python
new_df
```

```python
# code to replace element bsed of columns.
new_df = df.replace({
    'temprature': -99999,
    'windspeed': [-99999, -88888],
    'event': '0'
}, np.NaN)
```

```python
new_df
```
