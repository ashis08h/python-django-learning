```python
# code to convert day column data in to day time.
df = pd.read_csv('data_folder/newyork1.csv', parse_dates=['day'])
```

```python
df
```

```python
type(df['day'][0])
```

```python
# coe to set the index of dataframe as day column.
df.set_index('day', inplace=True)
```

```python
df
```

```python
# code to fill all the na value with 0 entire dataframe.
new_df = df.fillna(0)
```

```python
new_df
```

```python
# code to replace NA value with some other value for a particular column.
new_df = df.fillna({
    'temprature': 0,
    'windspeed': 0,
    'event': 'No event'
})
```

```python
new_df
```

```python
# code to fill the na value with just forward value means one value above.
new_df = df.fillna(method='ffill')
```

```python
new_df
```

```python
# code to copy next day value
new_df = df.fillna(method='bfill')
```

```python
new_df
```

```python
# code to copy data from next column at the place of na
new_df = df.fillna(method='bfill', axis='columns')
```

```python
new_df
```

```python
# same things.
new_df = df.fillna(method='bfill', axis=1)
```

```python
new_df
```

```python
# we can add limit also while copying data from previous or next to NA.
new_df = df.fillna(method='ffill', limit=1)
```

```python
new_df
```

```python
# code to interpolate the data means it will insert data between previous and next, linear interpolation.
new_df = df.interpolate()
```

```python
new_df
```

```python
# code to interpolate in terms of time.
new_df = df.interpolate(method='time')
```

```python
new_df
```

```python
# code to drop the row having atleast one na values from the dataframe.
new_df = df.dropna()
```

```python
new_df
```

```python
# code to drop the row having all the na element.
new_df = df.dropna(how='all')
```

```python
new_df
```

```python
# code to keep that row having atleast one non na value else drop.
new_df = df.dropna(thresh=1)
```

```python
new_df
```

```python
# code to keep the row having atleast two non na value else drop.
new_df = df.dropna(thresh=2)
```

```python
new_df
```

```python
# code to add those rows which are missing like day 2nd and 3rd
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
