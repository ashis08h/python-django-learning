```python
# code to merge two data frame into one excelsheet having different sheet.
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})
```

```python
df_stocks
```

```python
df_weather = pd.DataFrame({
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temprature': [32, 35, 28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('../../data_folder/s1/stocks_weather.xlsx') as writter:
    df_stocks.to_excel(writter, sheet_name='stocks', index=False)
    df_weather.to_excel(writter, sheet_name='weather', index=False)
```
