```python
df = pd.read_csv('../../data_folder/new_weather.csv')
```

```python
# code to replace -99999 with NaN value
df.replace(-99999, np.NaN)
```

```python
# code to replace multiple special values with NaN
df.replace([-99999, -88888], np.NaN)
```

```python
# code to replace the special values based on the columns.
df.replace({'temprature': -99999,
           'windspeed': -88888,
           'event': '0'}, np.NaN)
```

```python
df = pd.read_csv('../../data_folder/new_weather_unit.csv')
```

```python
df
```

```python
# code to replace string data with empty value
df.replace('[A-Za-z]', '', regex=True)
```

```python
# code to replace the string data with empty only in the temprature and windspeed columns not in the event
df.replace({
    'temprature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
}, '', regex=True)
```

```python
df
```

```python
df.replace({
    'temprature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
}, '', regex=True, inplace=True)
```

```python
df
```
