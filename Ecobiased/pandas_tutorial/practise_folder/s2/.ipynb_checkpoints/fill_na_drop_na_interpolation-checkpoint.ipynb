```python
# code to fill with 0 in the entire dataframe inplace of null value
df.fillna(0)
```

```python
df
```

```python
# code to fill with different values to the different columns 
df.fillna({
    'temprature': 0,
    'windspeed': 0,
    'event': 'No event'
})
```

```python
df
```

```python
# code to replace the na value with one value forward.
df.fillna(method='ffill')
```

```python
df
```

```python
# code to replace the na value with next value
df.fillna(method='bfill')
```

```python
# code to replace na value from the next column
df.fillna(method='bfill', axis=1)
```

```python
# same things
df.fillna(method='bfill', axis='columns')
```

```python
# we can add limit also while replacing na value from next or previous
# here limit 1 means 0 and 1 will start counting with zero
df.fillna(method='ffill', limit=1)
```

```python
df
```

```python
# code to replace the na value with a value which lies between prev value and next val
# this will only apply on numeric values.
df.interpolate()
```

```python
# code to interplate interms of time
df.interpolate(method='time')
```

```python
# code to drop those rows having atleast one null value
df.dropna()
```

```python
# code to drop the rows having all the value as null
df.dropna(how='all')
```

```python
# code to keep those rows having atleast one non null value else drop it
df.dropna(thresh=1)
```

```python
df.dropna(thresh=2)
```
