```python
# write a 3 by 3 dataframe and add one column at last and store the value of 20 percent of previous column
# if value of previous column is greater than 35
```

```python
df = pd.DataFrame({
    'name': ['Ashish', 'Kumar', 'Sah'],
    'Age': [28, 23, 24],
    'Salary': [100, 200, 300]
})
```

```python
df
```

```python
df['result'] = df['Salary'].apply(lambda x: x*0.2 if x < 200 else None)
```

```python
df
```

```python
# need to create a new column name as result and the value will be 20 % of salary if age of those row are less than 24 else None
```

```python
def calculate_result(row):
    if row['Age'] < 24:
        return row['Salary']*0.5
    else:
        return None

df['Result'] = df.apply(calculate_result, axis=1)
```
